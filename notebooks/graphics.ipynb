Used to create graphics during the capstone

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
Prices= pd.read_csv('../data/Prices.csv', encoding='utf-8',thousands=',')
Cross_Border = pd.read_csv('../data/Cross_Border.csv',encoding='utf-8',thousands=',')
Consumption = pd.read_csv('../data/Cosumption.csv',encoding='utf-8',thousands=',')
Generation = pd.read_csv('../data/Generation.csv',encoding='utf-8',thousands=',')
```

The columns Start date and End date instead of datatype datetime datatype object. The reason is in the csv format itself. Therefore:

```python
Prices['Start date'] = pd.to_datetime(Prices['Start date'])
Prices['End date'] = pd.to_datetime(Prices['End date'])

Cross_Border['Start date'] = pd.to_datetime(Cross_Border['Start date'])
Cross_Border['End date'] = pd.to_datetime(Cross_Border['End date'])

Consumption['Start date'] = pd.to_datetime(Consumption['Start date'])
Consumption['End date'] = pd.to_datetime(Consumption['End date'])

Generation['Start date'] = pd.to_datetime(Generation['Start date'])
Generation['End date'] = pd.to_datetime(Generation['End date'])
```

```python

start_date = '2022-12-20'
end_date = '2023-01-7'


filter_gen = Generation[(Generation['Start date'] >= start_date) & (Generation['Start date'] < end_date)]

col = list(Generation.columns)[2:-1]
colors = [
    "#003f5c",
    "#2f4b7c",
    "#665191",
    "#a05195",
    "#d45087",
    "#f95d6a",
    "#ff7c43",
    "#ffa600",
    "#ffcc00",
    "#ff9900",
    "#cc6600",
    "#993300"
]

fig, ax1 = plt.subplots(figsize=(15, 6), dpi=200)

filter_gen.set_index('Start date')[col].plot(kind='area', stacked=True, ax=ax1, color=colors)


ax1.set_title('Energy generation and Prices from ' + start_date + ' to ' + end_date)
ax1.set_ylabel('Energy in MWh')
ax1.set_xlabel('Time')




ax1.legend(title='Sources', bbox_to_anchor=(1.36, 1.014), loc='upper right')

plt.show()
```

```python
import matplotlib.pyplot as plt

start_date = '2022-12-20'
end_date = '2023-01-07'

filter_cons = Consumption[(Consumption['Start date'] >= start_date) & (Consumption['Start date'] < end_date)]


fig, ax = plt.subplots(figsize=(15, 6), dpi=200)

ax.plot(filter_cons['Start date'], filter_cons['Total (grid load) [MWh] Original resolutions'], color='black', label='Total Consumption', linewidth=2.5)

ax.set_title('Total demand between ' + start_date + ' to ' + end_date)
ax.set_ylabel('Demand in MWh')
ax.set_xlabel('Time')
plt.xticks(rotation=45)
ax.legend(loc='upper right')


plt.tight_layout()
plt.show()
```
